```python
#As we can see column Status and unnamed don't contain any value, so they will not play any part analysis. Delete it
df.drop(['Status','Unnamed'], axis= 1, inplace= True)
```

```python
#in amount column there are some null values, check for null
pd.isnull(df)
#this function show is there is null value or not in the form of true and false. but as we can see it is not helpful
```

```python
#we will use the sum function with isnull to get a better idea
pd.isnull(df).sum()
#pd.isnull(df) find the null value and sum() add all the null value to show us a clear picture
```

```python
#one way to deal with the null value is we can simply drop them with df.dropna(inplace=true) or we can fill the missing value.
```

```python
#filter rows with at least one null value
rows_with_null = df[df.isnull().any(axis=1)]
#here we are finding the rows which contains the null values that is why we are using 'any'
# print("Rows with null values:")
print(rows_with_null)
```

```python
#calculate the mean for each category and fill NA with mean for that category
df['Amount'] = df.groupby('Product_Category')['Amount'].transform(lambda x: x.fillna(x.mean()))
#here we are saying python to fill the null value in the amount column with the mean value of the product_category.
#groupby means collection of same thing, The transform function in pandas DataFrame allows for element-wise operations.
"""This lambda function, x.fillna(x.mean()), performs an operation on a pandas Series x.
x is a pandas Series that represents the values in the 'Amount' column of each group within your DataFrame, where each group is defined by 'Product_Category'.
x.fillna(x.mean()) calculates the mean of all non-NA/null values in the Series x and fills any NA/null values with this mean value. This essentially replaces any missing values in the 'Amount' column of each group with the mean of the non-missing values in that group.
The lambda function is used here because it allows you to create a quick, inline function without explicitly defining it using def. In this case, the lambda function takes one argument (x) and returns x.fillna(x.mean()). This lambda function is applied to each group of data in the 'Amount' column within your DataFrame."""
```

```python
#define a function to categorize age
def categorize_age(age):
    if age <= 17:
        return '0-17'
    elif age <= 20:
        return '18-25'
    elif age <= 35:
        return '26-35'
    elif age <= 45:
        return '36-45'
    elif age <= 50:
        return '46-50'
    elif age <= 55:
        return '51-55'
    else:
        return '55+'
# Apply the function to create a new column 'age_group'
df['Age_Group'] = df['Age'].apply(categorize_age)
print(df)
        
```

# EDA - Exploratory Data Analysis

### - Gender

```python
# plotting a bar chart for gender and it's count with data label
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)
bx=sns.barplot(x= 'Gender', y= 'Amount' , data = sales_gen)

# for bars in bx.containers:
#     bx.bar_label(bars)
```

From above graphs we can see that most of the buyers are females and even thee purchasing power of female are greater than men

```python
cx = sns.countplot(data = df , x = 'Age_Group', hue = 'Gender')

for bars in cx.containers:
    cx.bar_label(bars)
```

```python
cx = sns.countplot(data = df , x = 'Age_Group')
```

```python
#total amount vs age group
sales_age = df.groupby(['Age_Group'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.barplot(x = 'Age_Group', y = 'Amount' , data = sales_age)
```

From above graphs we can see that most of the buyers are of age group betwen 26 - 35 years female

```python
#total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data = sales_state, x = 'State', y = 'Orders')
```

```python
#total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data = sales_state, x = 'State', y = 'Amount')
```

From above graphs we can see that most of the orders & total sales/amount are from UP, Maharashtra and karnataka respectively.
